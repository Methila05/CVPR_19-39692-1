```python
import matplotlib.pyplot as plt
import numpy as np
  
x = np.arange(-20, 20, 2)
print(x)
```

Activition Function

```python
def plot_graph(y,ylabel):
    plt.figure()
    plt.plot(x,y, 'o--')
    plt.grid(True)
    plt.xlabel("X")
    plt.ylabel(ylabel)
    plt.show()
```

Step Function

```python
y = list(map(lambda n: 1 if n>0.5 else 0, x))
plot_graph(y,"STEP(X)")
```

Sigmoid Function

```python
y = 1 / (1 + np.exp(-x))
plot_graph(y, "Sigmoid(X)")
```

Tanh

```python
y = (np.exp(2*x) - 1) / (np.exp(2*x) + 1)
plot_graph(y, "TanH(X)")
```

Relu

```python
y = list(map(lambda a: a if a>=0 else 0, x))
plot_graph(y,"ReLU(X)")
```

Elu

```python
alpha = 1
y = np.where(x > 0, x, alpha * (np.exp(x) - 1))
plot_graph(y,"ELU(X)")
```

Selu

```python
lamda = 1.0507
alpha = 1.6734
y = np.where(x > 0, x, lamda*alpha * (np.exp(x) - 1))
plot_graph(y,"SeLU(X)")
```
